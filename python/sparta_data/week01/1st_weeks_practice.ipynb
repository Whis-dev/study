```python
chicken08 = pd.read_csv('./data/chicken_08.csv')
chicken09 = pd.read_csv('./data/chicken_09.csv')
chicken_data = pd.concat([chicken07, chicken08, chicken09])
chicken_data
```

```python
chicken_data = chicken_data.reset_index(drop=True)
chicken_data
```

```python
groupdata = chicken_data.groupby('요일')
call_data = groupdata['통화건수']
call_sum_data = call_data.sum()
call_sum_data
```

```python
plt.figure(figsize=(8,5))
plt.bar(call_sum_data.index, call_sum_data)
plt.title('요일별 통화건수의 합')
```

```python
print('설정 되어 있는 폰트 사이즈 :', plt.rcParams['font.size'])
print('설정 되어 있는 폰트 글꼴 :', plt.rcParams['font.family'])
```

```python
plt.figure(figsize=(8,5))
plt.bar(call_sum_data.index, call_sum_data)
plt.title('요일별 통화건수의 합')
```

```python
sorted_data = call_sum_data.sort_values(ascending=False)
sorted_data
```

```python
plt.figure(figsize=(8,5))
plt.bar(call_sum_data.index, sorted_data)
plt.title('요일별 통화건수의 합')
```

```python
weeks = ['월', '화', '수', '목', '금', '토', '일']
weeks_data = chicken_data.groupby('요일').sum()['통화건수'].reindex(weeks)
weeks_data
```

```python
plt.figure(figsize=(8,5))
plt.bar(weeks_data.index, weeks_data)
plt.title('요일별 통화건수의 합')
```

```python
age_data = chicken_data.groupby('연령대').sum()['통화건수']
age_data
```

```python
sorted_age_data = age_data.sort_values(ascending = False)
sorted_age_data
```

```python
plt.figure(figsize=(8,5))
plt.bar(sorted_age_data.index, sorted_age_data)
plt.title('연령대별 치킨 전체 주문량')
```
