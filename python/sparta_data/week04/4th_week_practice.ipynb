```python
code = pd.read_csv('./data/corpgeneral.csv', header=0)
code = code[['회사명', '종목코드']]
code_result = code.rename(columns={"회사명":"corp", "종목코드":"code"})
code_result
```

```python
def get_code(code_result,corp_name):
    condition = "corp=='{}'".format(corp_name)
    
    corp_code = code_result.query(condition)
    corp_code = corp_code['code']
    corp_code = corp_code.to_string(index=False)
    corp_code = corp_code.strip()
    corp_code = corp_code.rjust(6, "0")
    corp_code = corp_code + ".KS"
    return corp_code
```

```python
stock_code = get_code(code_result, '삼성전자')
stock_code
```

```python
companies = ['삼성전자', 'LG전자', '카카오', 'NAVER', 'CJ', '한화', '현대자동차', '기아자동차']
```

```python
start = datetime(2019, 1, 1)
end = datetime(2020, 9, 10)
```

```python
stock_of_companies = pd.DataFrame({'Date':pd.date_range(start=start, end=end)})
stock_of_companies
```

```python
for company in companies:
    code_item = get_code(code_result, company)
    stock_df = pdr.get_data_yahoo(code_item, start, end)
    stock_of_companies = stock_of_companies.join(pd.DataFrame(stock_df['Close']).rename(columns={'Close':company}), on="Date")

stock_of_companies
```

```python
corr_data = stock_of_companies.corr()
corr_data
```

```python
plt.figure(figsize=(5,3))
ax = sns.lineplot(data=kakao_stock['Close'])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=corr_data, annot=True, fmt='.2f', linewidth=.5, cmap='Blues')
```

```python
stocks_of_companies = pd.DataFrame({'Date': pd.date_range(start=start, end=end)})

for company in companies:
    company_code = get_code(code_result, company)
    stock_df = pdr.get_data_yahoo(company_code, start, end)
    stocks_of_companies = stocks_of_companies.join(pd.DataFrame(stock_df['Close'].pct_change()).rename(columns={'Close':company}), on='Date')

stocks_of_companies.dropna()
stocks_of_companies.to_csv('./stock_change.csv',sep=',', na_rep='NaN',index = False)
```
