```python
import pandas as pd
```

```python
commercial = pd.read_csv('./data/commercial.csv')
commercial
```

```python
commercial.tail(5)
```

```python
list(commercial)
```

```python
commercial_id = commercial.groupby('상가업소번호')['상권업종소분류명'].count()
commercial_id
```

```python
category = set(commercial['상권업종소분류명'])
category
```

```python
commercial['도로명주소']
```

```python
commercial[['시', '구', '상세주소']] = commercial['도로명주소'].str.split(' ', n = 2, expand = True)
commercial
```

```python
seoul_data = commercial[commercial['시'] == '서울특별시']
seoul_data
```

```python
set(seoul_data['시'])
```

```python
seoul_chicken_data = seoul_data[seoul_data['상권업종소분류명'] == '후라이드/양념치킨']
seoul_chicken_data
```

```python
gu_chicken_count = seoul_chicken_data.groupby('구').count()['상가업소번호']
gu_chicken_count
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"
```

```python
plt.figure(figsize = (20,5))
plt.bar(gu_chicken_count.index, gu_chicken_count)
plt.title('구별 치킨상가 갯수')
```

```python
gu_chicken_count = gu_chicken_count.sort_values(ascending = False)
```

```python
geo_data = './data/seoul_geo.json'
geo_data = json.load(open(geo_data, encoding='utf-8'))
geo_data
```

```python
map = folium.Map(location=[37.55,126.98], zoom_start=11, tiles='stamentoner')
map
```

```python
folium.Choropleth(
    geo_data=geo_data,
    data=gu_chicken_count,
    columns=[gu_chicken_count.index, gu_chicken_count],
    fill_color='PuRd',
    key_on="feature.properties.name").add_to(map)
```

```python
population_map = folium.Map(location=[37.55, 126.98], zoom_control=11, tiles='stamentoner')
population_map
```

```python
folium.Choropleth(geo_data=geo_data,
                 data=seoul_population_gu,
                 columns=[seoul_population_gu.index, seoul_population_gu],
                 fill_color='PuRd',
                 key_on='feature.properties.name').add_to(population_map)
```

```python
population_map
```

```python
gu_chicken_count = pd.DataFrame(gu_chicken_count).reset_index()
gu_chicken_count
```

```python
seoul_population_gu = pd.DataFrame(seoul_population_gu).reset_index()
seoul_population_gu
```

```python
gu_chicken = gu_chicken_count.join(seoul_population_gu.set_index('군구'), on='구')
gu_chicken
```

```python
gu_chicken['유동인구수/상가갯수'] = gu_chicken['유동인구수']/gu_chicken['상가업소번호']
gu_chicken
```

```python
plt.figure(figsize=(20,5))
gu_chicken = gu_chicken.sort_values(by='유동인구수/상가갯수')
plt.bar(gu_chicken['구'], gu_chicken['유동인구수/상가갯수'])
plt.title('구별 유동인구수/상가갯수')
```
